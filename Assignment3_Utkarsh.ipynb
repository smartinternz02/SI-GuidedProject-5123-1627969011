```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,25,35)
y=x**2
```

```python
plt.plot(x,y)
plt.xlabel('x-label')
plt.ylabel('x^2-values')
plt.title('line plot')
plt.grid(True)
plt.show()
```
